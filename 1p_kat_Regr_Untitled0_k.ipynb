```python
import pandas as pd

path="/content/drive/MyDrive/Dan/internship_train.csv"
dataframe = pd.read_csv(path)
```

```python
dataframe.info()
```

```python
dataframe.dropna(inplace=True)

dataframe = dataframe.astype(float)
```

```python
dataframe.head()
```

```python
dataframe.info()
```

```python
X=dataframe.iloc[:,:-1].values
#y=dataframe[['target']].values
y=dataframe.iloc[:,-1:].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
linr = LinearRegression()

linr.fit(X_train, y_train)
y_train_pred = linr.predict(X_train)
y_test_pred = linr.predict(X_test)
```

```python
path="/content/drive/MyDrive/Dan/internship_hidden_test.csv"
dataframe1 = pd.read_csv(path)
#dataframe1 = pd.read_csv('./internship_hidden_test.csv')
dataframe1.dropna(inplace=True)

dataframe1 = dataframe1.astype(float)
```

```python
dataframe1.head()
```

```python
X1 = dataframe1.iloc[:, :]

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score


print('MSE train: {:.3f}, test: {:.3f}'.format(
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```
